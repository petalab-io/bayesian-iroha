# 事前分布の比較シミュレーション
分析者が「データを分析する前」に持っている知識の強さを、3つのパターンで可視化しています。今回の売上予測モデルで `Normal(11, 2)` を選んだ意図を明確にします。

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import japanize_matplotlib

# パラメータ設定
x = np.linspace(-5, 15, 1000)

# 1. 無情報事前分布 (Flat/Uninformative): 何も知らない
prior_uninformative = stats.norm.pdf(x, 0, 100)

# 2. 弱情報事前分布 (Weakly Informative): 「この範囲ならあり得る」という常識
prior_weakly = stats.norm.pdf(x, 11, 2)

# 3. 強情報事前分布 (Informative): 「絶対にこうだ」という強い信念
prior_informative = stats.norm.pdf(x, 11, 0.2)

# 可視化
plt.figure(figsize=(12, 6))
plt.plot(x, prior_uninformative, label="無情報 (Flat): 予測がつかない", linestyle="--", color="gray")
plt.plot(x, prior_weakly, label="弱情報 (Weakly): 妥当な範囲を示す", linewidth=3, color="C0")
plt.plot(x, prior_informative, label="強情報 (Informative): 頑固な確信", linewidth=2, color="C1")

plt.title("事前分布のタイプ別比較 (売上の対数スケール想定)", fontsize=14)
plt.xlabel("パラメータの値 (例: 推定される売上の対数)", fontsize=12)
plt.ylabel("確率密度", fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
